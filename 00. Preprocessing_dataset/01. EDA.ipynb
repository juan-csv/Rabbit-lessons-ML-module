```python
import pandas as pd
import numpy as np
```

# Load dataset

```python
path_ds = "/Users/macbook/Desktop/Rabbit/census_dataset/census.csv"
df = pd.read_csv(path_ds)
print(f"Census Dataset len ------> {len(df)}")
df.head()
```

# Remove row's with Nan and Null values

```python
df = df.dropna()
```

# Explore categorical variables

```python
categorical_key = ["workclass", "native_country", "marital_status", "occupation", "race"]
for key_col in categorical_key:
    print(f"-------- {key_col} ----------")
    ls = list(df[key_col].unique())
    print(ls)
    #print(f"Number of unique values ------>  {len(ls)}")
    print("----------------------------------------------\n")
```

# One Hot Encoding

```python
for key_col in categorical_key:
    # one hot encoding
    one_hot = pd.get_dummies(df[key_col], prefix=key_col)
    # remove column
    df = df.drop(key_col,axis = 1)
    # add one hot encoding
    df = df.join(one_hot)
```

# Explore ordinal variables

```python
categorical_key = ["age", "education_num", "hours_per_week"]
for key_col in categorical_key:
    print(f"-------- {key_col} ----------")
    ls = list(np.sort(df[key_col].unique()))
    print(ls)
    #print(f"Number of unique values ------>  {len(ls)}")
    print("----------------------------------------------\n")
```

# Normalize ordinal columns

```python
for key_col in categorical_key:
    ls = np.sort(df[key_col].unique())
    min = ls.min()
    max = ls.max()
    normalize = lambda x: (x-min)/(max-min)
    df[key_col] = df[key_col].apply(normalize)
```

```python
# transform output
df["income_bracket"] = df["income_bracket"].apply(lambda x: 1 if x==" <=50K" else 0)
```

# Split dataset Train, Validation, Test

```python
#df_train
df_train = df[df["dataframe"]=="training"]
df_validation = df[df["dataframe"]=="evaluation"]
df_test = df[df["dataframe"]=="prediction"]
```

# Save Dataset

```python
df_train.to_csv("train_census.csv",index=False)
df_validation.to_csv("validation_census.csv",index=False)
df_test.to_csv("test_census.csv",index=False)
```
